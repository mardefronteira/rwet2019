```python
# importing counter
from collections import Counter
```

```python
# creating counter
corpoCount = Counter(corpoHoje)
```

```python
# making a list of all the words in the text, without repetitions. the extra "", [] and : are just to make my job easier
# in adding the phonemes by hand
corpoHoje = corpoCount.most_common()
# why does it work when I pop 0? shouldn't 0 be the string?
corpoHoje = sorted(['"'+list(item).pop(0)+'\" : [\"\"]' for item in corpoHoje])
corpoHoje
```
