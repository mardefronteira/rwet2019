> ## Version 1
The first version of Native Speaker takes three translations of the source text, and using the original language as a template, replaces each word randomly with the correspondent from one of the four texts.<br>Since the translations are not word by word, this correspondence is not related to the replaced word's translation, but to the index of the words in the poem. That way, the poem progressively dissolves its original meaning.

```python
import random
import textwrap

corpoHoje = open("corpoHoje.txt").read().split("-----")

langText = {
    'pt': corpoHoje[0].split(),
    'en': corpoHoje[1].split(),
    'fr': corpoHoje[2].split(),
    'es': corpoHoje[3].split()
}

texto = []

# selecting language
for num in range(len(langText['pt'])):
    i = random.randrange(20)
    
#   compose text using different languages
    
    if i <= 8:
        texto.append(langText['pt'][num])
    elif i > 8  and i <= 16:
        texto.append(langText['en'][num])
    elif i > 16 and i <= 19:
        texto.append(langText['fr'][num])
    else: 
        texto.append(langText['es'][num])
    
# uppercasing the initial letter if last word ends in punctuation
for palavra in texto:
    if palavra[-1] in '.!?' and texto.index(palavra) != len(texto) - 1:
        texto[texto.index(palavra)+1] = texto[texto.index(palavra)+1].title()
    elif texto.index(palavra) < len(texto)-1:
        texto[texto.index(palavra)+1] = texto[texto.index(palavra)+1].lower()

extraSpace = [63,60,59,58,55,54,53,52,50,49,48]
for i in extraSpace:
    texto[i] = "      "+texto[i]

lineBreaks = [278,279,268,269,241,242,231,232,184,185,186,182,183,184,179,172,163,154,145,138,135,130,127,125,120,102,103,83,84,81,66,67,63,60,59,58,55,54,53,52,50,49,48,39,40,30,25,19,15,16]
for i in lineBreaks:
    texto.insert(i, '\n')
    
print(" ".join(texto))
```

>### Version 2.0
Using editdist to compare the pronunciations for each word to the CMU dictionary used by pronouncingpy, the program randomly choses between the most similar pronunciation notations -- which can differ from the actual closest pronunciation -- and replaces the Brazilian words with similar sounding words in English. Unfortunately, having the CMU dictionary as the input to compare to the Brazilian sounds returns a big number of uncommon words, or words rooted in other languages, which makes it difficult to perceive it as English.

```python
# create empty dictionary to contain the distances
distDict = {}

# list to hold final composition
composition = []

# go through each word in the poem in order
for word in corpoHoje_words:
    distDict[word] = {}
    
    # compare it to every word in the cmu dictionary
    for phone in cmuDict:
        
        # calculate distances
        dist = editdistance.eval(corpoHoje_phones[word][0], cmuDict[phone])
        
        if dist < 15:
        
            # append to the closest words' dictionary, under the distance's key
            if dist in distDict[word]:
                distDict[word][dist].append(phone)
            else:
                distDict[word][dist] = [phone]
        
    for i in range(15):
        if i in distDict[word]:
            composition.append(random.choice(distDict[word][i]))
            break

# print(len(corpoHoje_words),len(composition),composition)
```

>### Version 2.1
